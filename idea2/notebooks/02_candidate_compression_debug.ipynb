# Candidate Compression Debug

```python
from binary_sanity_suite import oracle_logits, random_logits_like
from data import generate_lwe_batch
from metrics import evaluate_binary_candidates, random_candidate_hit_probability
from recovery import recover_binary_trace, trace_to_row
from utils import resolve_device, seed_everything

seed_everything(11)
device = resolve_device('cuda')
n, M, q, h, K = 16, 128, 127, 3, 8
batch = generate_lwe_batch(batch_size=16, n=n, M=M, q=q, h=h, sigma_e=0.0, device=device, secret_type='binary')
oracle = {'support_logits': oracle_logits(batch.s)}
random = {'support_logits': random_logits_like(batch.s)}
print('expected random hit:', random_candidate_hit_probability(n, h, K))
print('oracle:', evaluate_binary_candidates(batch, oracle, h, K))
print('random:', evaluate_binary_candidates(batch, random, h, K))
trace = recover_binary_trace(batch.A[0], batch.b[0], batch.s[0], oracle['support_logits'][0], h, K, q)
print(trace_to_row(trace))
```
